```python
from nltk.corpus.reader import WordListCorpusReader
reader= WordListCorpusReader("C:/Users/PS-STATE HOUSE/Desktop",
["engilishwords.txt"]) #creating corpus
```

```python
bx = reader.raw("engilishwords.txt")#reading corpus text
```

```python
punctuationRemoved = bx.replace(r'[^\w\s]+', '') #removing punctuation from text
```

```python
lower_case_corpus = punctuationRemoved.lower()#changing corpus to all lower case to remove conflict between upper case and lowercase version of the same word
```

```python
ax = word_tokenize(lower_case_corpus)#tokenizing lower case corpus by word
```

```python
ax
```

```python
punctuation=['"','#','$','%',"'",'(',')',',','*','+','-','/',':',';','<','>','=','@','[','\\',']','^','_','~','{','}','|','``']
final_words=[x for x in ax if x not in punctuation] #filtering more non word elements from text
```

```python
final_words
```

```python
listOfBigrams = []#list of Bigramss
listOfUnigrams = []#list of Unigrams
bigramCounts = {}#bigram dictionary
unigramCounts = {}#unigram dictionary

for i in range(len(final_words)):#for word in corpus
        if i < len(final_words) - 1:

            listOfBigrams.append((final_words[i], final_words[i + 1]))#append current word and word ahead of it

            if (final_words[i], final_words[i+1]) in bigramCounts:#if current word and word ahead of current word is already in bigram counts dictionary
                bigramCounts[(final_words[i], final_words[i + 1])] += 1 #increase bigram count by one
                print("----------Bigram Counts----------")
                print(bigramCounts)
            else:
                bigramCounts[(final_words[i], final_words[i + 1])] = 1 #else make new entry for bigram counts
                print("----------Bigram Counts----------")
                print(bigramCounts)
                
        listOfUnigrams.append((final_words[i]))#append current word 
        if final_words[i] in unigramCounts: #if current word is in unigram count dictionary already
            unigramCounts[final_words[i]] += 1 #increase unigram count by one
            print("----------Unigram Counts----------")
            print(unigramCounts)
        else:
            unigramCounts[final_words[i]] = 1 #else make a new entry for unigram count
            print("----------Unigram Counts----------")
            print(unigramCounts)
```

```python
vocab = set(final_words) #getting unique(unrepeated) words from text
```

```python
sentencebigramCounts = {}#initializing dictionary
bi_sentence = "the african continent has potential" #sentence for bigram count table
tokenized_sentence = word_tokenize(bi_sentence)#tokenizing sentence by word
for tword in tokenized_sentence: #for every word in sentence
    for word in vocab: #for every unique word in text
        if(tword, word) in bigramCounts: # if combination of sentence word and corpus word is in bigram counts dictionary
            sentencebigramCounts[tword , word] = bigramCounts[tword , word] #assign bigramCounts value
            
        else:
            sentencebigramCounts[tword , word] = 0 #else, assign zero because there is no such pairing
```

```python
from prettytable import PrettyTable
 
# Specify the Column Names while initializing the Table
myTable = PrettyTable(["  ", "the", "african", "continent", "has", "potential"])
 
# putting all relevant info into table and creating bigram counts table for the above sentence
myTable.add_row(["the", str(sentencebigramCounts['the', 'the']), str(sentencebigramCounts['the', 'african']), str(sentencebigramCounts['the', 'continent']), str(sentencebigramCounts['the', 'has']), str(sentencebigramCounts['the', 'potential'])])
myTable.add_row(["african", str(sentencebigramCounts['african', 'the']), str(sentencebigramCounts['african', 'african']), str(sentencebigramCounts['african', 'continent']), str(sentencebigramCounts['african', 'has']), str(sentencebigramCounts['african', 'potential'])])
myTable.add_row(["continent", str(sentencebigramCounts['continent', 'the']), str(sentencebigramCounts['continent', 'african']), str(sentencebigramCounts['continent', 'continent']), str(sentencebigramCounts['continent', 'has']), str(sentencebigramCounts['continent', 'potential'])])
myTable.add_row(["has", str(sentencebigramCounts['has', 'the']), str(sentencebigramCounts['has', 'african']), str(sentencebigramCounts['has', 'continent']), str(sentencebigramCounts['has', 'has']), str(sentencebigramCounts['has', 'potential'])])
myTable.add_row(["potential", str(sentencebigramCounts['potential', 'the']), str(sentencebigramCounts['potential', 'african']), str(sentencebigramCounts['potential', 'continent']), str(sentencebigramCounts['potential', 'has']), str(sentencebigramCounts['potential', 'potential'])])

print(myTable)#print table
```

```python
import numpy
# Function takes sentence as input and and returns sentence probability using bigram table(which is sentencebigramCounts dictionary)  
def get_probability(sentence, bigram_counts, unigram_counts):
    #tokenize input sentence by word
    tokenized_input = word_tokenize(sentence)
    
    
    probabilities = {} #dictionary for probability of each bigram
    mult_prob= [] #list of probabilities for all bigrams
    for i in range(len(tokenized_input)):
        if i < len(tokenized_input) - 1:
            
            sentence_unigram = (tokenized_input[i]) #unigram of ith word
            sentence_bigram = (tokenized_input[i], tokenized_input[i+1]) #bigram containing ith word and it's successor

        
            sentence_bigram_count = bigram_counts.get(sentence_bigram) # getting bigram count from sentencebigramCounts dictionary
            sentence_unigram_count = unigram_counts.get(sentence_unigram) #getting unigram count from unigramCounts dictionary
        
            probability = sentence_bigram_count / sentence_unigram_count #calculating bi-gram probability accordingly
            probabilities[tokenized_input[i],tokenized_input[i+1]] = probability #updating dictionary
            mult_prob.append(probabilities[tokenized_input[i],tokenized_input[i+1]]) #appending probabilty to mult_prob
    
    mult_prob = list(mult_prob) #turning mult_prob to list
    result = numpy.prod(mult_prob) #using numpy to multiply all values in mult_prob
    print("Sentence Probability:", result) #printing probabilty of sentence
    return probabilities #return dictionary of probabilities
```
